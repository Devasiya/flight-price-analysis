```python
#importing basis 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#get the train and test dataset
train_df=pd.read_excel('Data_Train.xlsx')
test_df=pd.read_excel('Test_set.xlsx')
```

```python
train_df.head()
```

```python
#concat both the train and test datasets ,so that cleaning and eda and fearture extraction can be done together ,instead of doing it twice
df = pd.concat([train_df, test_df], ignore_index=True)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
#eda is done extensively in perviously ,now lets focus on feature engineering
df.columns
```

```python
# all the attributes are str ,so
#Date_of_Journey can be divided into day, month, year
df['Date_of_Journey'].str.split('/')
```

```python
df.shape
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
#now as date ,month,year is added ,lets remove Date_of_Journey
df=df.drop('Date_of_Journey',axis=1)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
#as we can see date,month,year ka datatype is object , lets us convert to int
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
#lets now move on to next feature that is arrival_time
df['Arrival_Time'].str.split(' ')
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df.info()
```

```python
#i forgot to check for null values ,which should be done in EDA
df.isnull().sum()
#there in no much null values
```

```python
#Arrival_Time can be further be divided into hour and min
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute']=df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
#drop arrival_time
df=df.drop('Arrival_Time',axis=1)
```

```python
df.info()
```

```python
#covert the datatype
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Minute']=df['Arrival_Minute'].astype(int)
```

```python
df.info()
```

```python
#lets do the same for Dep_Time
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Min']=df['Dep_Time'].str.split(':').str[1]
df['Dep_Hour']=df['Dep_Hour'].astype(int)
df['Dep_Min']=df['Dep_Min'].astype(int)
df=df.drop('Dep_Time',axis=1)
df.head(1)
```

```python
#now lets see for route ,which is unneccesary as we have source , destination and no of stops
```

```python
df.columns
```

```python
#dropping route col
df=df.drop('Route',axis=1)
```

```python
#lets move on to duration
#convert duration which is written in hour and min into min only
df['Duration_Hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration_Min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_Min'].isnull().sum()
```

```python
df['Duration_Min']=df['Duration_Min'].fillna(0)
```

```python
df['Duration_Min']=df['Duration_Min'].astype(int)
```

```python
#lets move to airline
df['Airline'].unique()
```

```python
#lets do label encoding for them 
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df['Airline']=labelencoder.fit_transform(df['Airline'])
df['Source']=labelencoder.fit_transform(df['Source'])
df['Destination']=labelencoder.fit_transform(df['Destination'])
df['Additional_Info']=labelencoder.fit_transform(df['Additional_Info'])
#here label encoding is done ,u can also do one hot encoding
```
